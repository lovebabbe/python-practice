```python
ip = advertisting.drop('Sales',axis=1)
op = advertisting.Sales
```

```python
ip.head()
```

```python
op.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(ip,op,test_size=0.2,random_state=30)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(pred,y_test)
r2 = r2_score (pred,y_test)
print('Mean Square Error: ',mse)
print('R2 Score: ',r2)
```

```python
df = pd.DataFrame({'Y_test':list(y_test),'Prediction':pred})
df
```

```python
sns.lmplot(x='Y_test',y='Prediction',data=df)
plt.show()
```
