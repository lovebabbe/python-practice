```python
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
cv2.imshow('Gray',gray)
cv2.waitKey(1000)
cv2.destroyAllWindows()
```

```python
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
blur = cv2.GaussianBlur(rgb_img,(7,9),20)
plt.imshow(blur)
plt.show()
```

```python
# edge detection

edge = cv2.Canny(rgb_img,6,100)
plt.imshow(edge,cmap='gray')
plt.show()
```

```python
#Capturing image through webcam
cap = cv2.VideoCapture(0)
if cap.isOpened():
    ret,frame = cap.read()
    print(ret)
    print(frame)
else:
    ret = False

image = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title('ye dosti')
plt.axis('off')
plt.show()
cap.release()
```

```python
# Video
vdo = r"E:\WhatsApp Video 2025-06-05 at 16.00.26_1b801442.mp4"

cv2.namedWindow(vdo)
cap = cv2.VideoCapture(vdo)

if cap.isOpened():
    ret, frame = cap.read()
    print(ret)
    print(frame)
else:
    ret = False
while ret:
    ret, frame = cap.read()
    if not ret:
        break
    cv2.imshow(vdo, frame)
    if cv2.waitKey(30) == ord('q'):
        break
cv2.destroyAllWindows() 
cap.release()
```
