```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

x = [20,30,45,67,78,39]
y = [30,25,35,16,38,34]

plt.plot(x,y,color='m')
plt.title('Line Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.grid()
plt.show()
```

```python
x = [20,30,45,67,78,39]
y = [30,25,35,16,38,34]

plt.plot(x,y,'*-',linewidth=2,markersize=13,color='y')
plt.title('Line Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.grid()
plt.show()
```

```python
x = [20,30,45,67,78,39]
y = [30,25,35,16,38,34]

plt.plot(x,y,'*-.',linewidth=2,markersize=13,color='y')
plt.title('Line Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.grid()
plt.show()
```

```python
x = [20,30,45,67,78,39]
y = [30,25,35,16,38,34]

plt.plot(x,y,'H-.',linewidth=2,markersize=13,color='b')
plt.title('Line Plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.grid()
plt.show()
```

```python
x = [20,30,45,67,78,39]
y = [30,25,35,16,38,34]

plt.bar(x,y,label='Boys',width=2,color='c')
plt.title('Bar Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.legend()
plt.show()
```

```python

x = [20, 30, 45, 67, 78, 39]
y = [30, 25, 35, 16, 38, 34]

plt.bar(x, y, label='Boys', width=5, color='c')
plt.plot(x, y, 'bh--', linewidth=2, markersize=10, color='orange')

plt.title('Bar and Line Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.annotate('Value', xy=(x[0], y[0]), xytext=(x[0], y[0]+5))

plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

counts = [15, 20, 10, 5]
flavors = ['Vanilla', 'Chocolate', 'Strawberry', 'Others']
colors = ['#f5a623', '#50e3c2', '#b8e986', '#d0021b']
plt.pie(counts, labels=flavors, colors=colors, shadow=True,
explode=(0.1, 0.1, 0.1, 0.1), autopct='%1.1f%%')
plt.title('Favorite Ice Cream Flavors')
plt.show()
```

```python
#Pie Chart

slices=[8,10,4,3,5]
activities=['sleeping', 'phone', 'study', 'eat', 'play']
col = ['gold', 'b', 'magenta', 'red', 'g']
plt.pie(slices,labels=activities, colors=col, startangle=98, explode=(0,0,0.1,0,0), shadow=True, autopct='%1.2f%%')
plt.title('Pie Chart')
plt.show()
```
