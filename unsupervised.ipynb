```python
from sklearn.feature_extraction.text import TfidfVectorizer,TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
```

```python
model = Pipeline([('vector',TfidfVectorizer()),
                 ('tfidf',TfidfTransformer()),
                 ('algo',MultinomialNB())])
```

```python
model.fit(news_train.data,news_train.target)
```

```python
def predict_category(s,tr=news_train,m=model):
    pred = m.predict([s])
    return tr.target_names[pred[0]]
```

```python
predict_category('audi is better tha hyundai')
```

```python
predict_category('he is our prime minister')
```
